##### Q5.C
Write a function that accepts one parameter, a list
The function should return the sum total of all numbers in the list, using a loop.
\
*Hint*: Do we want a for loop or a while loop for this kind of question?

```python
# your code here

def sum_list(numbers):
    total = 0
    for num in numbers:
     total += num 
    return total

# A for loop in Python is designed specifically to iterate over a collection (like a list) from the first item to the last.
# It automatically handles moving to the next item and stops when the list ends, making your code clean and preventing accidental infinite loops.
```

```python
my_list=[3,7,1,9]

sum_list(my_list)
```

#### Q6. Problem-Solving Functions

```python
# A. Write a function that finds the longest string in a given list of strings

# your code here

def longest_string_finder(strings):
    if not strings:
        return ""
    
    longest = strings[0]
    
    for s in strings:
        if len(s) > len(longest):
            longest = s
            
    return longest
```

```python
s=["dfaf","adfaffaf","grrtsrfsgfgfsgsgf","adfaaa"]

longest_string_finder(s)
```

#### Q7. Challange Question (Optional)  

```python
# A. Write a function to find the most common letter in a given string, and use the previous function in your answer
#    Note: do not include the space " " as a character

# your code here
def char_count(text, letter):
    return text.count(letter)

def most_common_letter(text):
    if not text or text.isspace():
        return None
        
    most_frequent_char = None
    highest_count = 0
    
    for char in text:
        if char == " ":
            continue
            
        current_count = char_count(text, char)
        
        if current_count > highest_count:
            highest_count = current_count
            most_frequent_char = char
            
    return most_frequent_char
```

```python
s3="This is a test about data science short course with excellent students"

most_common_letter(s3)
```
